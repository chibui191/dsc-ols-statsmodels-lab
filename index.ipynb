## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries 
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore')
```

```python
# import the data
df = pd.read_csv('advertising.csv')
```

```python
# Check the columns and first few rows
df.head()
```

```python
# Drop column 'Unnamed: 0'
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset
df.info()
```

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20,5))
for i in range(4):
    sns.distplot(df[df.columns[i]], ax=axes[i])
fig.show();
```

```python
# Record your observations here
# TV and radio look symmetrical, although not perfectly bell-shaped
# Sales seems to be normally distributed, slightly right-skewed
# Newspaper seems to be normally distributed, heavily right-skewed
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
channels = df.columns[:3]
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(21,7))
for i in range(3):
    sns.scatterplot(df[channels[i]], df.sales, ax=axes[i])
fig.show();
```

```python
# Record yor observations on linearity here 
# TV and radio seem to have a linear relationship with sales, 
# newspaper does not show a clear linear relationship with sales --> might violate linearity assumption
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# build the formula 
f_TV = 'sales ~ TV'

# create a fitted model in one line
model = ols(formula=f_TV, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

```python
# R-Square is 0.612, meaning roughly 61% of the variation of the target is explained by the model
# 1 "unit" increase in TV spending only corresponds with 0.0475 "unit" increase in sales
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
x_TV = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
# make predictions for those x values and store them
y_TV_pred = model.predict(x_TV)
```

```python
# first, plot the observed data and the least squares line
fig, ax = plt.subplots(figsize=(10, 8))
sns.scatterplot(df.TV, df.sales, label="Observed Data", color='black', ax=ax)
sns.lineplot(x_TV.TV, y_TV_pred, label='Regression Line', color='red', ax=ax)
ax.legend()
fig.show();
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, 'TV', fig=fig)
plt.show();
```

```python
# Record Your observations on heteroscedasticity
# From the Residuals vs TV plot: cone-shaped --> heteroscedasticity
# --> violate the homoscedasticity assumption
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
residuals = model.resid
fig = sm.graphics.qqplot(residuals,
                         dist=stats.norm,
                         line='45',
                         fit=True)
plt.show();
```

```python
# Record Your observations on the normality assumption

# Q-Q plot looks good, as data appear as to follow along the diagonal line
# --> seems to be normally distributed
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
# formula for radio
f_radio = 'sales ~ radio'
# create a fitted model
model = ols(formula=f_radio, data=df).fit()
```

```python
# create a DataFrame for min & max values of df.radio
x_radio = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
y_radio_pred = model.predict(x_radio)
```

```python
# plot the observed data and the least squares line
fig, ax = plt.subplots(figsize=(10, 8))
sns.scatterplot(df.radio, df.sales, label="Observed Data", color='black', ax=ax)
sns.lineplot(x_radio.radio, y_radio_pred, label='Regression Line', color='red', ax=ax)
ax.legend()
fig.show();
```

```python
fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, 'radio', fig=fig)
plt.show();
```

```python
residuals = model.resid
fig = sm.graphics.qqplot(residuals,
                         dist=stats.norm,
                         line='45',
                         fit=True)
fig.show();
```

```python
model.summary()
```

```python
# Record your observations here for goodnes of fit 

# From the Residuals vs radio plot: cone-shaped --> heteroscedasticity
# --> violate the homoscedasticity assumption

# Q-Q plot also looks off, data does not closely follow along the diagonal line
# --> does not fulfill the normality assumption

# R-Square is 0.332, meaning only 33% of the variation is explained by the model
# 1 "unit" increase in radio spending corresponds with 0.2025 "unit" increase in sales

# Skewness & Kurtosis seem higher than TV

# --> Overall, radio does not perform as well as TV as a predictor
```

There is obvious heteroscdasticity in the residuals for both variables.
